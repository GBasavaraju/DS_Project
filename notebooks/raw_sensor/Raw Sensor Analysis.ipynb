```python
import numpy as np

import heartpy as hp
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import config

from analysis import read_all_dfs, filter_outliers_signal, plot_sensor_values, filter_frequencies, show_bpm
```

```python
def do_analysis(analysis_type, col):
    dfs = read_all_dfs(config.DATA_FOLDER, analysis_type, config.DATA_FILES[analysis_type])
    for label, _, df in dfs:
        filtered_df = filter_outliers_signal(df.copy(), col)
        fs = hp.get_samplerate_mstimer(df['TIME'].values)
        print('Sample Rate:', fs)
        plot_sensor_values(filtered_df[col], label, 'Heart Beats Raw')
        filtered_signal = filtered_df[col].values
        filtered_signal = hp.flip_signal(filtered_signal)
        filtered_signal = filter_frequencies(filtered_signal, fs)
        plot_sensor_values(filtered_signal[2000:3000], label, 'Heart Beats Filtered', logarithmic=True)
        show_bpm(filtered_signal, fs)
```

# Idle Analysis

```python
do_analysis('idle', 'IR')
```

# Exercise Analysis

```python
do_analysis('after_plank', 'IR')
```
